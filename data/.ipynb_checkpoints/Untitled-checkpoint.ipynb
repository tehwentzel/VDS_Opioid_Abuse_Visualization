```python
new=df[['pat_id','pharmacynpi',"physiciannpi","days_supply"]]
```

```python
from collections import defaultdict  # available in Python 2.5 and newer

pha_count= dict()
phy_count=dict()
pharisphy=list()
count=0
for index, row in new.iterrows():
    if not str(row.pharmacynpi) in pha_count:
        pha_count[str(row.pharmacynpi)]=1
    else:
        pha_count[str(row.pharmacynpi)]+=1
    if not str(row.physiciannpi) in phy_count:
        phy_count[str(row.physiciannpi)]=1
    else:
        phy_count[str(row.physiciannpi)]+=1
 
    if(row.pharmacynpi==row.physiciannpi):
        pharisphy.append(1)
        count=count+1
    else:
        pharisphy.append(0)
```

```python
node_size={'pharmacy':pha_count,'physician':phy_count}
```

```python
import json
myjson = json.dumps(node_size)
f = open("node_size.json","w")
f.write(myjson)
f.close()
```

```python
new['pharisphy']=pharisphy
```
